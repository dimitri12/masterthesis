```python
import pandas as pd
df = pd.read_csv("testdata_exjobb.csv", encoding='ISO-8859-1')
df
```

```python
#dummies = pd.get_dummies(df.FreeText)
dummies = df.iloc[:,0:2]
dummies
```

```python
import re
sentence = [None] * df.shape[0]
for i in range(len(sentence)):
    old_sentence = dummies.iloc[i].FreeText
    word = list(old_sentence.split())
    words = [None] * len(word)
    for i in range(len(word)):
        words[i] = re.sub('(?<=\D)[.,]|[.,](?=\D)', '', word[i].lower())
    words1 = [x for x in words if x is not None]
    sentence.append(' '.join(words1))
    sentence1 = [x for x in sentence if x is not None]
df['newFreeText'] = [str(x) for x in sentence1]
df
```

```python
from numpy import array
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from numpy import argmax
words3=[]
nsentence = [None] * df.shape[0]
for i in range(len(nsentence)):
    old_sentence = dummies.iloc[i].FreeText
    word = list(old_sentence.split())
    words = [None] * len(word)
    for i in range(len(word)):
        words[i] = re.sub('(?<=\D)[.,]|[.,](?=\D)', '', word[i].lower())
    words2 = [x for x in words if x is not None]
    words3.append(words2)
words3
```

```python
import gensim
from gensim.models import Word2Vec
from gensim.models import word2vec
from gensim.models import Phrases
import logging
```

```python
model = Word2Vec(words3, min_count=1)
# summarize the loaded model
print(model)
# summarize vocabulary
words4 = list(model.wv.vocab)
print(words4)
# access vector for one word
print(model['bäckasiner'])
```

```python
from sklearn.decomposition import PCA
from matplotlib import pyplot
```

```python
X = model[model.wv.vocab]
pca = PCA(n_components=2)
result = pca.fit_transform(X)
```
